```python
import pandas as pd
```

```python
df = pd.read_csv('orders.csv')
df['Time'] = pd.to_datetime(df['Time'], unit='ms')
```

```python
def highlight_us(val):
    if val == 'orchy':
        return 'background-color: yellow'
    elif val == 'Das Toona Fish':
        return 'background-color: aqua'
    elif val in ['Das Pacito Fish', 'Das Penguin Fish', 'Das Talon Fish']:
        return 'background-color: pink'
    elif val == 'Das Toxic Fish':
        return 'background-color: lavender'
    elif val == 'spoonily':
        return 'background-color: lime'
    elif val == 'Das Cow Fish':
        return 'background-color: orange'
    return ''
```

```python
df.set_index('Time').style.applymap(highlight_us)
```
